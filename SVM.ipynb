```python
cols_to_drop = [
    "id",
    "Unnamed: 32",
    "texture_se",
    "symmetry_se",
    "fractal_dimension_mean",
    "smoothness_se",
    "fractal_dimension_se"
]

df.drop(columns=cols_to_drop, inplace=True)
```

```python
df["Compactness_Ratio"] = df["compactness_mean"] / (df["smoothness_mean"] + 1e-6)


df["Area_Perimeter_Ratio"] = df["area_mean"] / (df["perimeter_mean"] + 1e-6)


df["Radius_Diff"] = df["radius_worst"] - df["radius_mean"]
```

```python
sample = df.head(50)

plt.figure(figsize=(8,4))
plt.plot(sample.index, sample["radius_mean"], label="Radius Mean")
plt.plot(sample.index, sample["radius_worst"], label="Radius Worst")

plt.title("Radius Mean vs Radius Worst")
plt.xlabel("Samples")
plt.ylabel("Radius")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create 3 categories
df["Category"] = "A"

df.loc[df["radius_mean"] >= 12, "Category"] = "B"
df.loc[df["radius_mean"] >= 18, "Category"] = "C"

# Plot
plt.figure(figsize=(7,5))

for category, color in zip(["A", "B", "C"], ["blue", "green", "red"]):
    subset = df[df["Category"] == category]
    plt.scatter(
        subset["radius_mean"],
        subset["texture_mean"],
        color=color,
        label=category,
        alpha=0.7,
        edgecolors="black"
    )

plt.title("Radius Mean vs Texture Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.legend(title="Category")
plt.grid(alpha=0.3)

plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt


le = LabelEncoder()
df['Category_encoded'] = le.fit_transform(df['Category'])


X = df[['radius_mean']]
y = df['Category_encoded']


model = SVC(kernel='linear')
model.fit(X, y)

score = float(input("Enter Radius Mean: "))


prediction_encoded = model.predict([[score]])
predicted_category = le.inverse_transform(prediction_encoded)[0]

print("Predicted Category:", predicted_category)


plt.figure(figsize=(8,5))


category_colors = {"A": "blue", "B": "green", "C": "red"}


for cat, color in category_colors.items():
    subset = df[df["Category"] == cat]
    plt.scatter(
        subset["radius_mean"],
        subset["texture_mean"],
        label=cat,
        alpha=0.7,
        edgecolors="black"
    )


representative_texture_mean = df[df["Category"] == predicted_category]["texture_mean"].mean()

# Plot the entered score as a new point
plt.scatter(
    score,
    representative_texture_mean,
    color="purple",
    marker="*",
    s=300,
    label=f"Your Input ({predicted_category})"
)

# Add text label for the entered score
plt.text(
    score,
    representative_texture_mean + 0.5,
    f"Predicted: {predicted_category}",
    fontsize=10,
    ha='center', va='bottom',
    bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="b", lw=1, alpha=0.6)
)

plt.title(f"Category Prediction based on Radius Mean and Texture Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.legend(title="Category")
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


columns_to_drop_from_X = ["diagnosis", "Category", "Category_encoded"]


X = df.drop(columns=columns_to_drop_from_X, axis=1)
y = df["diagnosis"]


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create SVM Model
svm_model = SVC(
    kernel='rbf',
    C=1.0,
    gamma='scale',
    random_state=42
)

# Train the Model
svm_model.fit(X_train, y_train)

# Predictions
y_pred = svm_model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("SVM Accuracy:", accuracy)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
